```python
import pandas as pd
import numpy as np
import re
import string 
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data_fake = pd.read_csv(r'E:\downloads\Fake .csv')
data_true = pd.read_csv(r'E:\downloads\True.csv') 
```

```python
data_fake['class'] = 0  
data_true['class'] = 1 
```

```python
data_combined = pd.concat([data_fake, data_true])
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(f'[{re.escape(string.punctuation)}]', '', text)
    return text
```

```python
data_combined['text'] = data_combined['text'].apply(preprocess_text)
```

```python
X = data_combined['text']
y = data_combined['class']
```

```python
vectorizer = CountVectorizer()
X_vectorized = vectorizer.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub(f'[{re.escape(string.punctuation)}]', '', text)
    return text
```

```python
def manual_testing(news):
    # Create a dictionary for the input news article
    testing_news = {"text": [news]}
    
    # Create a DataFrame from the dictionary
    new_def_test = pd.DataFrame(testing_news)
    
    # Apply the wordopt function to preprocess the text
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    
    # Transform the processed text using the vectorizer (make sure vectorizer is defined)
    new_x_test = vectorizer.transform(new_def_test["text"])
    
    # Make prediction using your trained model (make sure model is defined)
    prediction = model.predict(new_x_test)
    
    if prediction[0] == 0:
        return "The news is predicted as FAKE."
    else:
        return "The news is predicted as TRUE."
```

```python
# Input from user for testing 
news_input = str(input("Enter a news article: "))
result = manual_testing(news_input)
print(result)
```
